```python
from Data import data_processing as dp
from sklearn.metrics import precision_score, recall_score
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, classification_report
from Baseline.baseline_analysis import analyze_baseline_model
```

```python
X_train, X_test, y_train, y_test, vectorizer = dp.get_training_data()
```

```python
# Create a baseline model that predicts the most frequent label
baseline_model = DummyClassifier(strategy='most_frequent')

# Fit the baseline model to the training data
baseline_model.fit(X_train, y_train)

# Make predictions on the test set using the baseline model
y_pred = baseline_model.predict(X_test)
```

```python
print(y_pred)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
print(f"Baseline Model Test Accuracy: {accuracy}")
print("Classification Report for Test Set:")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(classification_report(y_test, y_pred, zero_division=0))
```
